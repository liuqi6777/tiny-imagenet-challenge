```python
import torch
from PIL import Image
from torchvision import transforms
from datasets import load_dataset
```

```python
data_samples = load_dataset('Maysee/tiny-imagenet')['train'].select(range(10))
```

```python
img = data_samples[2]['image']
img
```

```python
img.save('figs/origin.png')
```

```python
aug1 = transforms.RandomHorizontalFlip(p=1)(img)
aug1.save('figs/hflip.png')
aug1
```

```python
aug2 = transforms.RandomResizedCrop(224)(img)
# aug2.save('figs/crop.png')
aug2
```

```python
aug3 = transforms.Grayscale()(img)
aug3.save('figs/gray.png')
aug3
```

```python
aug4 = transforms.ColorJitter(brightness=[0.8, 0.8])(img)  # 亮度
aug4.save('figs/brightness.png')
aug4
```

```python
aug5 = transforms.ColorJitter(contrast=[1.4, 1.4])(img)  # 对比度
aug5.save('figs/contrast.png')
aug5
```

```python
aug6 = transforms.ColorJitter(saturation=[1.4, 1.4])(img)  # 饱和度
aug6.save('figs/saturation.png')
aug6
```

```python
transforms.ColorJitter(hue=[0.1, 0.1])(img)  # 色调
```

```python
aug7 = transforms.GaussianBlur(7)(img)
aug7.save('figs/gaussianblur.png')
aug7
```
